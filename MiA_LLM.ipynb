```python
print("Loading GPT2...")
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
tokenizer.padding_side = "left"
tokenizer.pad_token = tokenizer.eos_token

model1 = GPT2LMHeadModel.from_pretrained('gpt2-xl', return_dict=True).to('cuda')
model1.config.pad_token_id = model1.config.eos_token_id
```

```python
input_text = "What the fuck did you just fucking say to me, you little bitch?"
input_ids = tokenizer.encode(input_text, return_tensors="pt").to('cuda')
```

```python
import torch
with torch.no_grad():
    output = model1.generate(input_ids, max_length=500,eos_token_id=tokenizer.eos_token_id)
```

```python
generated_text = tokenizer.decode(output[0])
print(generated_text)
```

```python
import torch
print("Loading GPT2...")
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
tokenizer.padding_side = "left"
tokenizer.pad_token = tokenizer.eos_token

model1 = GPT2LMHeadModel.from_pretrained('gpt2', return_dict=True).to('cuda')
model1.config.pad_token_id = model1.config.eos_token_id
input_text2 = "Jone Doe lives in [], England"
input_id2s = tokenizer.encode(input_text2, return_tensors="pt").to('cuda')

with torch.no_grad():
    output = model1.generate(input_id2s, max_length=500, eos_token_id=tokenizer.eos_token_id)
generated_text = tokenizer.decode(output[0])
print("First output:", generated_text)
```
